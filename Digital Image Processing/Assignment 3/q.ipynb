```python
# read Given Lena image
img = cv2.imread('./Lena.png')

# converting BGR to RGB
img = cv2.cvtColor(img, cv2.COLOR_BGR2RGB)

# plotting the image using matplotlib
plt.imshow(img)
plt.title('Original Image')
plt.show()


# printing the original image shape
print('Original Image Shape: ', img.shape)
```

1. Convert the given Lena image to grayscale image. Use the cv2.resize() to down sample
the image with 4 sizes (128x128 , 64x64, 32x32, and 16x16). Display the original image,
and down sampled images with the same display size. Observe what happens.

```python
# converting the image to grayscale
gray = cv2.cvtColor(img, cv2.COLOR_RGB2GRAY)

# plotting the grayscale image
plt.imshow(gray, cmap='gray')
plt.title('Grayscale Image')
plt.show()
```

```python
sizes = [128, 64, 32, 16]

# Down-sample the image with different sizes
images = [cv2.resize(gray, (size, size)) for size in sizes]

# Plot the original image and the down-sampled images
fig, ax = plt.subplots(1, len(images) + 1, figsize=(15, 5))
ax[0].imshow(gray, cmap="gray")
ax[0].set_title("Original Image")
for i in range(len(images)):
    ax[i+1].imshow(images[i], cmap="gray")
    ax[i+1].set_title("Size: {}x{}".format(sizes[i], sizes[i]))
plt.show()
```

### Conclusion
The down-sampled grayscale images are likely showing unclear pixels because of a process called "Aliasing". This is when the information lost in downsampling causes the image to appear jagged or have a "staircase" effect. This is a common issue with images that have been down-sampled and can result in loss of detail and clarity in the image. Aliasing is an effect that causes different signals/pixels to become indistinguishable when sampled.







2. Down sample the grayscale Lena image with 8 different intensity ranges of values (0-
255, 0-128, 0-64, 0-32, 0-16, 0-8, 0-4, and 0-2). (Note: Size of images are the same).
And display all those 8 downsampled images in the same size display area on the
screen. Observe what happens.

```python
# Define the intensity ranges
ranges = [128, 64, 32, 16, 8, 4, 2, 1]

intensity_ranges = [255, 127, 63, 31, 15, 7, 3, 1]

# Down-sample the image with different intensity ranges
images = [gray//range for range in ranges]

# Plot the down-sampled images
fig, ax = plt.subplots(1, len(images), figsize=(15, 5))
for i in range(len(images)):
    ax[i].imshow(images[len(images) - i - 1], cmap="gray")
    ax[i].set_title("Range: 0-{}".format(intensity_ranges[i]))
plt.show()
```

### Conclusion

Reducing the range of intensity values in the image results in loss of detail and smoother appearance due to quantization of the grayscale values.

Decreasing the gray-level resolution of a digital image may result in what is known as false contouring. This effect is caused by the use of an insufficient number of gray levels in smooth areas of a digital image.
