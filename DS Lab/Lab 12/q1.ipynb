```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
m = [10, 100, 500, 1000, 5000, 10000, 50000]
data = []
for i in m:
    data.append(np.random.rand(i))

# print(data)
```

(a) Exponential Distribution

```python
for i in m:
    sample = np.random.exponential(1, i)
    bin = np.arange(0,10,0.1)
    plt.hist(sample, bins=bin, edgecolor='blue') 
    plt.title("Exponential Distribution for size = " + str(i)) 
    plt.show()
    # sns.kdeplot(sample, fill=True, color='xkcd:lightish blue')
```

(b) Uniform Distribution 

```python
for i in m:
    sample = np.random.uniform(0,1, i)
    bin = np.arange(0,1,0.1)
    plt.hist(sample, bins=bin, edgecolor='blue') 
    plt.title("Uniform Distribution for size = " + str(i)) 
    plt.show()
    # count, bins, ignored = plt.hist(sample, 15, density=True)
    # plt.plot(bins, np.ones_like(bins), linewidth=2, color='r')
    # plt.show()
```

(c) Bernoulli Distribution

```python
for i in m:
    sample = np.random.binomial(1,0.5, i)
    bin = np.arange(0,1,0.1)
    plt.hist(sample, bins=bin, edgecolor='blue') 
    plt.title("Bernoulli Distribution for size = " + str(i)) 
    plt.show()
```

(a) Exponential Distribution

```python
d = {}
for i in m:
    sample = np.random.exponential(0.33, i)
    d[str(i)] = (np.mean(sample))

#print(d)

plt.bar(list(d.keys()), list(d.values()), color ='cyan',width = 0.4)
 
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.title("Exponential Distribution")
plt.show()
# mean = expectation = 1/lambda = 0.33

    
```

 (b) Uniform Distribution

```python
for i in m:
    sample = np.random.uniform(0,1, i)
    d[str(i)] = (np.mean(sample))

#print(d)

plt.bar(list(d.keys()), list(d.values()), color ='cyan',width = 0.4)
 
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.title("Uniform Distribution")
plt.show()

# mean = expectation = 0+1/2 = 0.5
```

(c) Bernoulli Distribution

```python
for i in m:
    sample = np.random.binomial(1,0.5, i)
    d[str(i)] = (np.mean(sample))

#print(d)

plt.bar(list(d.keys()), list(d.values()), color ='cyan',width = 0.4)
 
plt.xlabel("Sample Size")
plt.ylabel("Sample Mean")
plt.title("Bernoulli Distribution")
plt.show()

# mean = expectation = p = 0.5
```

The Weak Law of Large Numbers, also known as Bernoullis theorem, states that if you have a sample of independent and identically distributed random variables, as the sample size grows larger, the sample mean will tend toward the population mean.

As the sample size n grows to infinity, the probability that the sample mean x-bar differs from the population mean mu by some small amount epsilon is equal to 0.
